```python
import gradio as gr
from transformers import TFAutoModelForSeq2SeqLM, AutoTokenizer

# Load the pre-trained model and tokenizer
model_name = "t5-base"  # You can choose any other model as well
model = TFAutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Define a function for text summarization
def summarize_text(input_text):
    inputs = tokenizer.encode("summarize: " + input_text, return_tensors="tf", max_length=512, truncation=True)
    summary_ids = model.generate(inputs, max_length=150, min_length=40, length_penalty=2.0, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

# Define input and output components
input_textbox = gr.Textbox(lines=10, label="Input Text")
output_textbox = gr.Textbox(label="Summary")

# Create a Gradio interface for the function
iface = gr.Interface(
    fn=summarize_text,
    inputs=input_textbox,
    outputs=output_textbox
)

# Launch the interface
iface.launch()

```

```python
import gradio as gr
from transformers import TFAutoModelForCausalLM, AutoTokenizer

# Load the pre-trained model and tokenizer for GPT-2
model_name = "gpt2"  # You can choose any other GPT-2 variant as well
model = TFAutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Define a function for text generation
def generate_continuation(input_text):
    inputs = tokenizer.encode(input_text, return_tensors="tf", max_length=512, truncation=True)
    output_ids = model.generate(inputs, max_length=150, num_return_sequences=1, early_stopping=True)
    generated_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)

    # Remove the input text from the generated output
    generated_text = generated_text[len(input_text):].strip()

    return generated_text

# Define input and output components
input_textbox = gr.Textbox(lines=10, label="Input Text")
output_textbox = gr.Textbox(label="Generated Continuation")

# Create a Gradio interface for the function
iface = gr.Interface(
    fn=generate_continuation,
    inputs=input_textbox,
    outputs=output_textbox
)

# Launch the interface
iface.launch()
```

```python
import gradio as gr
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the pre-trained model and tokenizer for GPT-2
model_name = "gpt2"  # You can choose any other GPT-2 variant as well
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Define a function for text summarization
def summarize_text(input_text):
    input_ids = tokenizer.encode(input_text, return_tensors="pt", max_length=512, truncation=True)
    output_ids = model.generate(input_ids, max_length=150, num_return_sequences=1, early_stopping=True)
    summary = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return summary

# Define input and output components
input_textbox = gr.Textbox(lines=10, label="Input Text")  # Changed from gr.inputs.Textbox to gr.Textbox
output_textbox = gr.Textbox(label="Summary")

# Create a Gradio interface for the function
iface = gr.Interface(
    fn=summarize_text,
    inputs=input_textbox,
    outputs=output_textbox,
    title="Text Summarization Chatbox",
    description="Enter a text and get a summary!",
    theme="compact"
)

# Launch the interface
iface.launch()
```

```python
import gradio as gr
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Load the pre-trained model and tokenizer for GPT-2
model_name = "gpt2"  # You can choose any other GPT-2 variant as well
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Define a function for question answering
def answer_question(question):
    input_text = "Question: " + question + "\nAnswer:"
    input_ids = tokenizer.encode(input_text, return_tensors="pt", max_length=512, truncation=True)
    output_ids = model.generate(input_ids, max_length=150, num_return_sequences=1, early_stopping=True)
    answer = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return answer

# Define input and output components
input_textbox = gr.inputs.Textbox(lines=3, label="Ask a question")
output_textbox = gr.outputs.Textbox(label="Answer")

# Create a Gradio interface for the function
iface = gr.Interface(
    fn=answer_question,
    inputs=input_textbox,
    outputs=output_textbox,
    title="Question and Answer Chatbox",
    description="Ask a question and get an answer!",
    theme="compact"
)

# Launch the interface
iface.launch()
```
